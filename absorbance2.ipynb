```python
import pandas as pd

# import csv data, 
df1 = pd.read_csv('./data/absorbance-2/NL0ML.CSV',sep = ',', encoding='utf-16')
df2 = pd.read_csv('./data/absorbance-2/NL1ML.CSV',sep = ',', encoding='utf-16')
df3 = pd.read_csv('./data/absorbance-2/NL2ML.CSV',sep = ',', encoding='utf-16')
df4 = pd.read_csv('./data/absorbance-2/NL125ML.CSV',sep = ',', encoding='utf-16')
```

```python
# display plot of data
plt1 = df1.plot(x='Wavelength (nm)', y='Absorbance (AU)', kind='line', figsize=(10,5), color='blue')
plt2 = df2.plot(x='Wavelength (nm)', y='Absorbance (AU)', kind='line', figsize=(10,5), color='purple', ax=plt1)
plt3 = df3.plot(x='Wavelength (nm)', y='Absorbance (AU)', kind='line', figsize=(10,5), color='#FFBE0B', ax=plt2)
plt4 = df4.plot(x='Wavelength (nm)', y='Absorbance (AU)', kind='line', figsize=(10,5), color='#FB5607', ax=plt3)

plt4.legend(["0 mL of KBr", "1 mL of KBr","2 mL of KBr","1.25 mL of KBr"]);
plt4.set_ylabel("Absorbance (A)")

```

```python
plt4.figure.savefig('absorbance-2.png')
```
